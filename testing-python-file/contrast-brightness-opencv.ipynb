```python
import os
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
path_img_good = '.\\src\\test\\1621237163.jpg'
path_img_bad = '.\\src\\test\\1626599086.jpg'
path_img_equalized = ".\\src\\test\\equalized.jpg"
```

```python
def histogramReturn (path):
    img = cv.imread(path, 0)
#     print("Image array - ", img)
    hist,bins = np.histogram(img.flatten(),256,[0,256])

    cdf = hist.cumsum()

    cdf_normalized = cdf * float(hist.max()) / cdf.max()
    plt.plot(cdf_normalized, color = 'b')
    plt.hist(img.flatten(),256,[0,256], color = 'r')
    plt.xlim([0,256])
    plt.legend(('cdf','histogram'), loc = 'center')

    plt.show()

    img = cv.imread(path, 1)
    plt.imshow(img)
```

```python
histogramReturn(path_img_bad)
```

```python
def histogramEqualize(img_code):
    path = f".\\src\\test\\{img_code}.jpg"
    img = cv.imread(path,0)                                 #Raw Image
    equ = cv.equalizeHist(img)                              #Equalized Image
    
    corrected_path = f'contrast_corrected\\contrast_corrected_{img_code}.png'
    cv.imwrite(corrected_path,equ)
    
    res = np.hstack((img,equ))                              #stacking images side-by-side
    side_path = f"contrast_corrected\\side_by_side_{img_code}.png"
    cv.imwrite(side_path,res)                              #Saves Image
    plt.imshow(cv.imread(side_path, 1))
```

```python
histogramEqualize(1626599072)
```
